```python
# main.py

from gen import GEN
import numpy as np

def funcion_test(x):
    return - (0.1 + ((1 - x) ** 2) - 0.1 * np.cos(6 * np.pi * (1 - x))) + 2 # 


menu = GEN(funcion_test, 
           population=300, 
           cant_genes=8, 
           num_cycles=100, 
           selection_percent=0.6, 
           crossing= 0.5, 
           mutation_percent=0.3, 
           i_min=0, 
           i_max=2,
           optimum='max'
           )

best_individual = menu.alg_stn_bin_rank()

print("best individual:", best_individual)

```

```python
# Ahora probamos el algoritmo cuántico
from gen import GEN
import numpy as np

def funcion_test_quantum(x):
    return - (0.1 + ((1 - x) ** 2) - 0.1 * np.cos(6 * np.pi * (1 - x))) + 2 #


menu = GEN(funcion_test_quantum,
           population=50,
           num_qubits=16,
           num_cycles=10,
           mutation_percent=0.05,
           i_min=-10,
           i_max=10
)

solucion = menu.alg_quantum()
print("Solución óptima (cuántico):", solucion)
```

```python
# Graficamos la función
import matplotlib.pyplot as plt
x = np.linspace(0, 2, 400)
y = funcion_test(x)
plt.plot(x, y)
plt.scatter(best_individual, funcion_test(best_individual), color='red')  # Punto óptimo encontrado
plt.title('Función de prueba y punto óptimo encontrado')    
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid()
plt.show()
```
